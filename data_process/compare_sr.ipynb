```python
def save_plot(figure, file_path, dpi=300, format='png'):
    """
    Save a Matplotlib figure to a file.

    Args:
        figure (matplotlib.figure.Figure): The Matplotlib figure to be saved.
        file_path (str): The path to save the figure.
        dpi (int): Dots per inch for image resolution (default is 300).
        format (str): The image file format (default is 'png').

    Returns:
        None
    """
    figure.savefig(file_path, format=format, dpi=dpi, bbox_inches='tight')
```

```python
import librosa
import librosa.display
import numpy as np

def wav_info(file, sr):
    # Load the audio file
    data, sr = librosa.load(file, sr=sr)
    print(file)
    print(f"data length: {len(data)}, sampling rate: {sr}")
    ipd.display(ipd.Audio(data, rate=sr)) # wav 파일 재생
    
    # Create a figure and a set of subplots
    fig,ax = plt.subplots(1,3,figsize=(12,3))
    
    # Waveform
    librosa.display.waveshow(data, sr=sr, ax=ax[0])
    
    # Mel spectrogram
    mel = librosa.feature.melspectrogram(y=data, sr=sr)
    librosa.display.specshow(librosa.power_to_db(mel, ref=np.max), y_axis='mel', x_axis='time', ax=ax[1])
    
    # MFCC
    mfcc = librosa.feature.mfcc(y=data, sr=sr)
    librosa.display.specshow(mfcc, x_axis='time', y_axis='mel', ax=ax[2])
    
    plt.tight_layout()
    plt.show()
```

```python
# 4000
num = np.random.randint(0, 3163, 3)
for file in np.array(files)[num]:
    wav_info(file, 4000)
```
